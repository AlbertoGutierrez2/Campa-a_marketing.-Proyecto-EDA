#### Analizamos la columna date

```python
print(f'El numero de valores nulos en la columna date es: {Bank_customer_details['date'].isnull().sum()}')


Bank_customer_details['date'].dtype
Bank_customer_details['date'].head(3)
```

```python
### Procedemos a cambiar el formato de la columna date que es de tipo objeto a tipo fecha. Para ello reemplazamos - por / y los meses en letra por 
### número de mes correspondiente.


from datetime import datetime

##Bank_customer_details['date'] = pd.to_datetime(Bank_customer_details['date'], format=  '%-d, %B, %Y')
##Bank_customer_details.head(3)

Bank_customer_details['date'] = Bank_customer_details['date'].astype(str)
Bank_customer_details['date'] = Bank_customer_details['date'].str.replace('-', '/')
Bank_customer_details['date'] = Bank_customer_details['date'].str.replace('enero', '01')
Bank_customer_details['date'] = Bank_customer_details['date'].str.replace('febrero', '02')
Bank_customer_details['date'] = Bank_customer_details['date'].str.replace('marzo', '03')
Bank_customer_details['date'] = Bank_customer_details['date'].str.replace('abril', '04')
Bank_customer_details['date'] = Bank_customer_details['date'].str.replace('mayo', '05')
Bank_customer_details['date'] = Bank_customer_details['date'].str.replace('junio', '06')
Bank_customer_details['date'] = Bank_customer_details['date'].str.replace('julio', '07')
Bank_customer_details['date'] = Bank_customer_details['date'].str.replace('agosto', '08')
Bank_customer_details['date'] = Bank_customer_details['date'].str.replace('septiembre', '09')
Bank_customer_details['date'] = Bank_customer_details['date'].str.replace('octubre', '10')
Bank_customer_details['date'] = Bank_customer_details['date'].str.replace('noviembre', '11')
Bank_customer_details['date'] = Bank_customer_details['date'].str.replace('diciembre', '12')


Bank_customer_details.head(3)


```

```python
### Una vez hechos los cambios anteriores convertimos la columna 'date' a datetime

Bank_customer_details['date'] = pd.to_datetime(Bank_customer_details['date'], format= 'mixed', dayfirst = True)
pd.set_option("display.date_dayfirst", True)
Bank_customer_details['date'].head(3)
```

```python
### Obtenemos la moda y las fechas mínima y máxima de la columna date. 

moda_fecha = Bank_customer_details['date'].mode()
print(f"La moda de la columna 'date' es: {moda_fecha}")

min_fecha = Bank_customer_details['date'].min()
print(f"El valor mínimo de la columna 'date' es: {min_fecha}")

max_fecha = Bank_customer_details['date'].max()
print(f"El valor máximo de la columna 'date' es: {max_fecha}")
```

```python
### Consideramos también generar una fecha aleatoria entre las fechas mínima y máxima de la columna 'date' pero finalmente obtamos por rellenar
### los 248 valores nulos con el valor de 0.

import random

ramdon_date = min_fecha + (max_fecha - min_fecha) * random.random()
ramdon_date

moda_fecha = Bank_customer_details['date'].mode()
Bank_customer_details['date'] = Bank_customer_details['date'].fillna(0)

Bank_customer_details['date'].isnull().sum()

```

#### Default, housing y loan

```python
### Para las variables numéricas default, housing y loan que recordamos lo que significan:
### default: Indica si el cliente tiene algún historial de incumplimiento de pagos (1: Sí, 0: No).
### housing: Indica si el cliente tiene un préstamo hipotecario (1: Sí, 0: No).
### loan: Indica si el cliente tiene algún otro tipo de préstamo (1: Sí, 0: No).
### Entendemos que lo más correcto es rellenar los nulos con ceros, porque si no tenemos conocimiento de si el cliente tiene historial de incumplimiento
### de pagos, un préstamo hipotecario o cualquier otro tipo de préstamo, lo más idóneo es indicar que no lo tiene hasta que se demuestre lo contrario.

Bank_customer_details['default'] = Bank_customer_details['default'].fillna(0)
Bank_customer_details['housing'] = Bank_customer_details['housing'].fillna(0)
Bank_customer_details['loan'] = Bank_customer_details['loan'].fillna(0)

Bank_customer_details['default'].isnull().sum()
Bank_customer_details['housing'].isnull().sum()
Bank_customer_details['housing'].isnull().sum()
```

```python
plt.figure(figsize = (10,2))
plt.title(f' Distribución de Age')

sns.boxplot(x= Bank_customer_details['age'])
plt.show()
```

```python
### Vemos en el box plot como nos encontramos con outliers o valores atípicos a partir de los 70 años y hasta los 98 años que es el valor máximo. 
### Estos valores se alejan del resto de datos del conjunto comprendidos entre los 33 y los 47 años aproximadamente y podrían estar falseando la
### la media, es decir incrementando la media (40 años) con respecto a la mediana (38 años). Por esa razón optamos por rellenar los nulos con el
### valor de la mediana y no de la media, ya que la mediana es más robusta y menos sensible que la media a valores atípicos.

mediana_age = Bank_customer_details['age'].median()
Bank_customer_details['age'] = Bank_customer_details['age'].fillna(mediana_age)

Bank_customer_details['age'].isnull().sum()



```
